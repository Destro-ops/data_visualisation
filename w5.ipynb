```python
for col in df.select_dtypes(include=['number']).columns: 
    mean_value = df[df[col] != -1][col].mean()  # Compute mean ignoring -1
    df[col] = df[col].apply(lambda x: mean_value if x==-1 else x)
df
```

```python
df.describe()   #The effectiveness of mean imputation depends on the dataset. If missing values are few and data is normally distributed, it’s acceptable. But if data is skewed or missing non-randomly, a different approach (e.g., median, regression) would be better.
```
