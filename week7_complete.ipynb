```python
#4

import matplotlib.pyplot as plt



venue_counts = df['venue'].value_counts()

# Plotting using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(venue_counts.index, venue_counts.values)
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.title('Number of Matches Held in Each Venue')
plt.xticks(rotation=-90)
plt.grid(True)
plt.show()



```

```python
#5
player_counts = df['player_of_match'].value_counts().head(10)
plt.bar(player_counts.index, player_counts.values, color='lightgreen')
plt.xlabel('Player')
plt.ylabel('Number of Player of the Match Awards')
plt.title('Top 10 Players with the Most Player of the Match Awards')
plt.xticks(rotation=45)
plt.show()
```

```python
#6
venue_counts=df["venue"].value_counts().tail(10)
plt.bar(venue_counts.index,venue_counts.values)
plt.xlabel("venue")
plt.ylabel("no. of games played")
plt.title("no. of games in stadiums")
plt.xticks(rotation=90)
plt.show()
```

```python
#7
win_counts=df.loc[df["venue"] == "M Chinnaswamy Stadium","winner"].value_counts()
win_counts
#sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))

plt.xticks(rotation=90)
sns.barplot(x=win_counts.index, y=win_counts.values, hue=win_counts.index, palette="magma", legend=False)

plt.title("Number of Wins at M Chinnaswamy Stadium", fontsize=16)
plt.xlabel("Winner", fontsize=12)
plt.ylabel("Win Count", fontsize=12)




#plt.show()
```

```python
#9


top_teams = df.groupby(['season', 'winner']).size().reset_index(name='wins')
top_teams=top_teams.sort_values(['wins', 'season'], ascending=[False, True])
top_teams = top_teams.groupby('season').head(5)
plt.figure(figsize=(14, 8))
sns.barplot(data=top_teams, x='season', y='wins', hue='winner', palette='viridis')

plt.title('Top 5 Teams (Season-wise) by Number of Wins', fontsize=16)
plt.ylabel('Number of Wins')
plt.xlabel('Season')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
#10

df['toss_win_match_win'] = df['toss_winner'] == df['winner']



toss_win_counts = df['toss_win_match_win'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=toss_win_counts.index,y=toss_win_counts.values,hue=toss_win_counts.index,legend=True)
plt.title("Relationship b/w winning toss and winning match", fontsize=12)
plt.ylabel("Matches")
plt.xlabel("outcome")
plt.xticks([0,1] , ['Loss after toss win' , 'Win after toss win'])
plt.legend(title='Legend', bbox_to_anchor=[1,1.025])
```
