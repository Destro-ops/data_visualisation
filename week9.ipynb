```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("sentiment.csv")
df
```

```python
#1
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Text'] = df['Text'].str.strip().str.lower()
df
```

```python
#2


sentiment_counts = df['Sentiment'].value_counts().head(10)
print(sentiment_counts)


plt.figure(figsize=(14, 14))
sentiment_counts.plot.pie(autopct='%1.1f%%')
plt.title('Sentiment Distribution')
plt.ylabel('')
plt.show()
```

```python
#3

text_length = df['Text'].str.len()
text_length_df = text_length.reset_index()
text_length_df.columns = ['Index', 'Text Length']
text_length_df



plt.figure(figsize=(10, 6))
sns.histplot(data=text_length_df, x='Text Length', bins=20, kde=True, color="blue")
plt.title("Distribution of Text Lengths")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()


```

```python
#6
df['Country'] = df['Country'].str.strip().str.title()

positive_sentiment_df = df[df['Sentiment'] == 'positive']
top_countries_positive = positive_sentiment_df['Country'].value_counts().head(5)
print(top_countries_positive)

plt.figure
sns.barplot(x=top_countries_positive.index, y=top_countries_positive.values, palette='crest')
plt.title('Top 5 Countries with Positive Sentiment')
plt.xlabel('Country')
plt.ylabel('Number of Posts')
plt.show()


```

```python
#10
df_filtered = df[df['Sentiment'].isin(['positive', 'negative', 'neutral'])]
year_country_sentiment = df_filtered.groupby(['Year', 'Country', 'Sentiment']).size().unstack(fill_value=0)
print(year_country_sentiment)

plt.figure(figsize=(14, 8))
sns.heatmap(year_country_sentiment.T, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)
plt.title('Sentiment Distribution by Year and Country')
plt.xlabel('Year & Country')
plt.ylabel('Sentiment')
plt.show()
```
